# CELLS TO RUN

Once the above are set, you can go a head and run the following cells

```python
# Load the desired files
sx_obj = sx.SparkXrif(obs_name, lab_name, sky_dark, lab_dark, sx.glob_dir_calib, sx.glob_mask, sx.glob_ref, sky_obs_n=obs_idx, dt_start=dt_start)
```

```python
lab_ref_sub = sx_obj.labref_rollsub
mask = sx_obj.mask_nan

# checking the lab reference: 
fig, axs = plt.subplots(nrows=1, ncols=4, figsize=(12,4), facecolor='white', sharey=True)
fig.subplots_adjust(wspace=0)

for e in range(4):
    axs[e].imshow(lab_ref_sub[e]*mask, vmax = np.max(lab_ref_sub), vmin = np.min(lab_ref_sub))
    axs[e].set_xlabel(f'ref {e} - ref {(e+1)%4}')

plt.suptitle(f"{obs_plot_name}: {sprk_type} \n GOOD DARKS ref sparkles, 1024 frames", y=0.90)
plt.savefig(obs_save_file + '_lab_ref_set')
```

```python
turb_file_lists = sx.gen_file_list(sx_obj.sky_obs_span)
clean_files, ts = sx_obj.file_sample_n_clean(turb_file_lists, 512*10, lab=False)
data_splt = np.array(sx.split_data(clean_files))
```

```python
# checking the on sky files:
fig, axs = plt.subplots(nrows=1, ncols=4, figsize=(12,4), facecolor='white', sharey=True)
fig.subplots_adjust(wspace=0)
plt.suptitle(f"On sky sparkles \n {obs_plot_name}: {sprk_type}", y=0.90)

sky_data = np.mean(data_splt, axis=1)

for e in range(4):
    axs[e].imshow(sky_data[e]*sx_obj.mask_nan, label = f'ref {e}',)
    axs[e].set_xlabel(f"ref {e}")

plt.savefig(obs_save_file + '_onsky_set')
```

```python
# Sum of all
plt.title(f"On sky sparkles, avg 10240 frames, GOOD REF \n {obs_plot_name}: {sprk_type} ", y=1.0)
plt.imshow(np.sum(data_splt, axis=(0,1))*sx_obj.mask_nan)
plt.savefig(obs_save_file + '_onsky_bg')
```

```python
# checking the on sky files:
fig, axs = plt.subplots(nrows=1, ncols=4, figsize=(12,4), facecolor='white', sharey=True)
fig.subplots_adjust(wspace=0)
plt.suptitle(f"On sky sparkles, Roll sub - Good REF  \n {obs_plot_name}: {sprk_type}", y=0.90)

sky_data = np.mean(data_splt, axis=1)

for e in range(4):
    e_i = (e+1)%4
    axs[e].imshow((sky_data[e] - sky_data[e_i])*sx_obj.mask_nan, label = f'ref {e}',)
    axs[e].set_xlabel(f"ref {e} - {e_i}")
plt.savefig(obs_save_file + '_onsky_bgsub_set')
```
